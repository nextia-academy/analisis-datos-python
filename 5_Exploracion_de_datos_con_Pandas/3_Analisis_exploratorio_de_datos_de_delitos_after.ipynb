#### 6. ¿En qué mes y en qué municipio ha ocurrido el mayor número de feminicidios?

```python
maxf = df[df.tipo_delito == 'Feminicidio'].frecuencia.max()
df[(df.tipo_delito == 'Feminicidio') & (df.frecuencia == maxf)]
```

```python
df[df.tipo_delito == 'Feminicidio'].sort_values(['frecuencia'], ascending=False).head()
```

#### 7. ¿En qué año y en qué estado ha ocurrido el mayor número de feminicidios?

```python
df[df.tipo_delito == 'Feminicidio'] \
    .groupby(['anio', 'entidad']) \
    .agg({'frecuencia':'sum'}) \
    .reset_index() \
    .sort_values(['frecuencia'],ascending=False) \
    .iloc[0]
```

#### 8. ¿Cuál es el municipio de la CDMX con mayor incidencia delictiva?

```python
df[df.clave_entidad == 9] \
    .groupby(['clave_municipio', 'municipio']) \
    .agg({'frecuencia':'sum'}) \
    .reset_index() \
    .sort_values('frecuencia', ascending=False) \
    .head(1)
```

```python
puebla = df[df.clave_entidad == 21]
xicotepec = puebla[puebla.municipio == 'Xicotepec']
hom_dol_x = xicotepec[xicotepec.subtipo_delito == 'Homicidio doloso'].frecuencia.sum()
hom_cul_x = xicotepec[xicotepec.subtipo_delito == 'Homicidio culposo'].frecuencia.sum()

plt.bar(['Homicidio Doloso', 'Homicidio Culposo'], [hom_dol_x, hom_cul_x])
```

### Ejercicio 2: Elige 3 estados de la república y grafica una serie de tiempo de la frecuencia abosluta de homicidios dolosos de enero 2015 a julio 2019 en estas tres entidades. Las tres líneas deberán estar en la misma figura.

```python
df_geo = pd.DataFrame({
    'clave_entidad': df.clave_entidad.unique(),
    'entidad': df.entidad.unique()
})
df_geo.head()
```

```python
df_geo[(df_geo.entidad == 'Oaxaca') | (df_geo.entidad == 'Ciudad de México') | (df_geo.entidad == 'Puebla')]
```

```python
df.head()
```

```python
df2 = df[(df.clave_entidad == 20) | (df.clave_entidad == 9) | (df.clave_entidad == 21) ]
df2 = df2[df2.subtipo_delito == 'Homicidio doloso']
df2.head()
```

```python
df2 = df2.groupby(['anio', 'mes', 'clave_entidad', 'entidad', 'subtipo_delito']) \
         .agg({'frecuencia':'sum'}) \
         .reset_index()
df2.head()
```

```python
df2['aniomes'] = df2.apply(lambda row: int(str(row['anio']) + str(row['mes']).zfill(2)), axis=1)
```

```python
df2.head()
```

```python
oax = df2[df2.clave_entidad == 20]
pue = df2[df2.clave_entidad == 21]
cdmx = df2[df2.clave_entidad == 9]
```

```python
oax.head()
```

```python
pue.head()
```

```python
plt.figure(figsize=(14,7))
plt.plot(range(len(cdmx.aniomes)), cdmx.frecuencia,'b-' ,label="CDMX" )
plt.plot(range(len(oax.aniomes)), oax.frecuencia,'r-' ,label="Oaxaca" )
plt.plot(range(len(pue.aniomes)), pue.frecuencia,'g-' ,label="Puebla" )
plt.xticks(range(len(cdmx.aniomes)), cdmx.aniomes, rotation=90)
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.plot(range(len(cdmx.aniomes)), cdmx.frecuencia,'b-' ,label="CDMX" )
plt.plot(range(len(oax.aniomes)), oax.frecuencia,'r-' ,label="Oaxaca" )
plt.plot(range(len(pue.aniomes)), pue.frecuencia,'g-' ,label="Puebla" )
plt.xticks(range(len(cdmx.aniomes)), cdmx.aniomes, rotation=90)
plt.legend(loc='best')
plt.show()
```
