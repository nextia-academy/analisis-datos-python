### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con estos valores

```python
df.Rating.unique()
```

```python
df.Rating.value_counts()
```

```python
ratings = df.Rating.value_counts()
type(ratings)
```

```python
ratings.index
```

```python
ratings.values
```

```python
df_ratings = pd.DataFrame({
    'rating' : ratings.index,
    'frequency' : ratings.values
})
df_ratings
```

### 22. Ordena el nuevo dataframe por la variable `rating`
